```python
#Refydo Muhammad Farhan
#12220059
import string
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import cm
```

```python
#Soal A
#Open File
file = pd.read_csv("produksi_minyak_mentah.csv")
#Change to Data Frame
df = pd.DataFrame(file, columns= ['kode_negara','tahun','produksi'])
country=input()
filter_negara=(df.loc[df['kode_negara']==country])
filter_negara
#Showing the plot
fig, ax = plt.subplots()
cmap = cm.get_cmap('tab10')
colors = cmap.colors
ax.set_title("Grafik Produksi Minyak Negara Setiap Tahun",fontsize=8)
ax.set_xlabel("Tahun")
ax.set_ylabel("Produksi")
ax.plot(filter_negara['tahun'], filter_negara['produksi'])
plt.locator_params(axis='x', nbins=45)
plt.xticks(rotation=90,fontsize = 6)
plt.xlim(1971,2015)
plt.tight_layout()
plt.show
```
